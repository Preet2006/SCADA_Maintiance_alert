```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)

# Load your labeled dataset
df = pd.read_csv("C:/Users/preet/OneDrive/Desktop/SHEATHING/labeled_data_v2.csv")
df['DATETIME'] = pd.to_datetime(df['DATETIME'])
```

```python
sns.countplot(data=df, x="FAILURE_LEVEL")
plt.title("Failure Level Distribution")
plt.xticks([0, 1, 2], ['Normal (0)', 'Watch (1)', 'Failure (2)'])
plt.show()
```

```python
correlation = df.select_dtypes(include='number').corr()['FAILURE_LEVEL'].sort_values()
correlation.plot(kind='barh')
plt.title("Correlation of Features with FAILURE_LEVEL")
plt.show()
```

```python
sns.boxplot(data=df, x="FAILURE_LEVEL", y="ACT SPEED")
plt.title("ACT SPEED Distribution by FAILURE_LEVEL")
plt.show()
```

```python
for label in df['FAILURE_LEVEL'].unique():
    plt.plot(df[df['FAILURE_LEVEL'] == label]['DATETIME'], 
             df[df['FAILURE_LEVEL'] == label]['ACT EXT RPM'], 
             label=f"Failure Level {label}", alpha=0.6)
plt.legend()
plt.title("ACT EXT RPM Over Time by Failure Level")
plt.xlabel("Time")
plt.ylabel("RPM")
plt.show()
```

```python
df['ZONE_1_TEMP_DIFF'] = df["ACT ZONE-1 TEMP"] - df["ZONE-1 TEMP"]
sns.boxplot(data=df, x="FAILURE_LEVEL", y="ZONE_1_TEMP_DIFF")
plt.title("ZONE-1 TEMP Deviation by FAILURE_LEVEL")
plt.show()
```
